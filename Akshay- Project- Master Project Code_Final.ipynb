```python
def plot_hexbin(dataframe, feature1, feature2):
    """
    Plot a hexbin plot for the given dataframe using the specified feature names.

    Parameters:
        dataframe (DataFrame): The pandas DataFrame containing the data.
        feature1 (str): The name of the first feature to plot on the x-axis.
        feature2 (str): The name of the second feature to plot on the y-axis.
    
    Returns:
        None
    """
    plt.figure(figsize=(10, 8))
    
    # Create hexbin plot
    plt.hexbin(dataframe[feature1], dataframe[feature2], gridsize=50, cmap='YlGnBu', mincnt=1)
    
    # Add colorbar
    plt.colorbar(label='Count')

    # Add linear regression line
    sns.regplot(x=dataframe[feature1], y=dataframe[feature2], scatter=False, color='orange', line_kws={'linewidth': 2})

    # Set title and labels
    plt.title(f'Hexbin Plot of {feature1} vs {feature2}')
    plt.xlabel(feature1)
    plt.ylabel(feature2)

    # Show plot
    plt.grid(True)
    plt.show()
```

```python
def plot_scatter_for_columns_containing_M(dataframe, datasetName):
    """
    Filters the predictive columns containing "M" in their names and creates pairwise scatter plots for the given dataframe.
    """
    # Filter columns containing "M" in their names
    filtered_columns = dataframe.filter(like='M')
    # Add the class column
    filtered_columns['Class'] = dataframe['Class']
    
    # Plot pairwise scatter plots
    sns.pairplot(filtered_columns, hue='Class', palette='viridis')
    plt.suptitle(f'Pairwise Scatter Plots of Columns containing "M" in {datasetName}', y=1.02)
    plt.show()
```

## NMF Analysis

```python
from sklearn.decomposition import NMF
import pandas as pd
```

```python
def apply_nmf(data, target_column='Class', n_components=2):
    """
    Apply Non-negative Matrix Factorization (NMF) to the given dataset.

    Parameters:
    - data (DataFrame): The input dataset.
    - target_column (str): The name of the target column.
    - n_components (int): The number of dimensions of the embedded space (default is 2).

    Returns:
    - DataFrame: DataFrame containing NMF-transformed features and target column.
    """

    # Extract features (X) and class labels (y)
    X = data.drop(target_column, axis=1)
    y = data[target_column]

    # Perform NMF
    nmf = NMF(n_components=n_components, random_state=42)
    X_nmf = nmf.fit_transform(X)

    # Create a DataFrame with NMF components and class labels
    nmf_df = pd.DataFrame(data=X_nmf, columns=[f'NMF{i+1}' for i in range(n_components)])
    nmf_df[target_column] = y

    return nmf_df
```

## Feature Selection

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.linear_model import LassoCV
from skrebate import ReliefF
```

```python
# define function to calculate the feature importnace using Different feature selection techniques
def calculateFeatureImportance(data, datasetName, k):
    print(f"----------------Feature Selection in dataset {datasetName}-------------------")
    # Extract features (X) and class labels (y)
    X = data.drop('Class', axis=1)
    y = data['Class']

    ### Using Tree Classifer for feature importance
    clf_rf = GradientBoostingClassifier(n_estimators=100, random_state=42)
    clf_rf.fit(X, y)
    feature_scores_rf = pd.Series(clf_rf.feature_importances_, index=X.columns).sort_values(ascending=False)

 
    ### ANOVA F-statistic
    selector_anova = SelectKBest(score_func=f_classif, k=k)
    selector_anova.fit(X, y)
    selected_feature_names_anova = X.columns[selector_anova.get_support()]
    # Sort selected features based on ANOVA F-statistic scores
    sorted_indices_anova = selector_anova.scores_[selector_anova.get_support()].argsort()[::-1]
    selected_feature_names_anova_sorted = selected_feature_names_anova[sorted_indices_anova]


    ### Mutual Information
    selector_mi = SelectKBest(score_func=mutual_info_classif, k=k)
    selector_mi.fit(X, y)
    selected_feature_names_mi = X.columns[selector_mi.get_support()]
    # Sort selected features based on Mutual Information scores
    sorted_indices_mi = selector_mi.scores_[selector_mi.get_support()].argsort()[::-1]
    selected_feature_names_mi_sorted = selected_feature_names_mi[sorted_indices_mi]


    ### Relief-based feature selection
    relief = ReliefF(n_features_to_select=k, n_jobs=-1)
    X_selected_relief = relief.fit_transform(X.values, y.values)
    selected_feature_names_relief = X.columns[relief.top_features_[:k]]
    
    # display the results
    print("\nSelected Features using Tree Classifier:", feature_scores_rf[:10].index)
    print("\nSelected features using ANOVA F-statistic:", selected_feature_names_anova)
    print("Selected features using Mutual Information:", selected_feature_names_mi)
    print("Selected features using ReliefF:", selected_feature_names_relief)

    ### Plotting the feature importances
    plt.figure(figsize=(15, 10))
    
    # Random Forest Feature Importances
    plt.subplot(2, 2, 1)
    sns.barplot(y=feature_scores_rf[:10].index, x=feature_scores_rf[:10], palette='viridis').set_title('Gradient Boosting Classifier Feature Importances')
    
    # ANOVA F-statistic
    plt.subplot(2, 2, 2)
    sns.barplot(y=selected_feature_names_anova_sorted, x=np.sort(selector_anova.scores_[selector_anova.get_support()])[::-1], palette='viridis').set_title('ANOVA F-statistic')

    # Mutual Information
    plt.subplot(2, 2, 3)
    sns.barplot(y=selected_feature_names_mi_sorted, x=np.sort(selector_mi.scores_[selector_mi.get_support()])[::-1], palette='viridis').set_title('Mutual Information')
    
    # ReliefF
    plt.subplot(2, 2, 4)
    sns.barplot(y=X.columns[relief.top_features_[:10]], x=sorted(relief.feature_importances_[:10],reverse=True), palette='viridis').set_title('ReliefF')

    plt.tight_layout()
    plt.show()


    return clf_rf.feature_importances_, selector_anova, selector_mi, relief
```

## Model Building and Evaluation

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier 
```

```python
# Import necessary libraries for report generation
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
```

```python
# Initialize the classifiers
def initialize_classifiers():
    classifiers = [
        ('Logistic Regression', LogisticRegression(max_iter=1000)),
        ('Quadratic Discriminant', QuadraticDiscriminantAnalysis()),
        ('K-Nearest Neighbors', KNeighborsClassifier()),
        ('Gradient Boosting', GradientBoostingClassifier(n_estimators=100, random_state=42)), 
        ('Multi-layer Perceptron', MLPClassifier(max_iter=1000, random_state=42))
    ]
    return classifiers
```

```python
def initialize_metrics():
    # Initialize dictionaries to store evaluation metrics
    accuracy_scores = {}
    precision_scores = {}
    recall_scores = {}
    f1_scores = {}
    confusion_matrices = {}
    return accuracy_scores, precision_scores, recall_scores, f1_scores, confusion_matrices
```

```python
def train_and_evaluate_models(classifiers):
    for name, classifier in classifiers:
        print(f'----Training {name} Model----')
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict(X_test)
        cm = confusion_matrix(y_test, y_pred)
        confusion_matrices[name] = cm
        accuracy_scores[name] = accuracy_score(y_test, y_pred)
        precision_scores[name] = precision_score(y_test, y_pred, average='binary')
        recall_scores[name] = recall_score(y_test, y_pred, average='binary')
        f1_scores[name] = f1_score(y_test, y_pred, average='binary')
    print('Models Training Complete.')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_mertics_line(classifiers, outputGraphName):
    # Define evaluation metrics
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    
    # Number of metrics
    num_metrics = len(metrics)
    
    # Custom colors for each classifier
    colors = plt.cm.get_cmap('tab10').colors
    
    # Compute x positions for each metric
    x_positions = np.arange(num_metrics)
    
    # Create line plot for each classifier
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot each classifier's metrics
    for i, (model, model_name) in enumerate(classifiers):
        values = [accuracy_scores.get(model), precision_scores.get(model), 
                  recall_scores.get(model), f1_scores.get(model)]
        
        ax.plot(x_positions, values, marker='o', color=colors[i], label=model_name)
    
    # Add labels and legends
    ax.set_xticks(x_positions)
    ax.set_xticklabels(metrics)
    plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=len(classifiers)//2)
    plt.title('Model Evaluation Metrics Line Plot')
    plt.xlabel('Metrics')
    plt.ylabel('Score')
    
    # Save and show plot
    plt.tight_layout()
    plt.savefig(f"{outputGraphName}_models_evaluation_line.png", dpi=300, bbox_inches='tight')
    plt.show()
    print(f"{outputGraphName}_models_evaluation_line.png successfully saved.")
```

```python
def generate_report(classifier_name, accuracy, precision, recall, f1, classifier):
    print(f"------- {classifier_name.upper()} Detailed Classification Report -------")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
    
    # Display classification report
    print("Classification Report:")
    print(classification_report(y_test, classifier.predict(X_test), target_names=["1","0"]))
```

```python
def plot_confusion_matrix(outputGraphName):
    plt.figure(figsize=(15, 10))
    for i, (name, cm) in enumerate(confusion_matrices.items(), 1):
        plt.subplot(2, 3, i)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix - {name}')
        plt.xlabel('Predicted Labels')
        plt.ylabel('True Labels')

    plt.tight_layout()
    plt.savefig(f'confusion-matix-{outputGraphName}.png',dpi=300,bbox_inches='tight')
    plt.show()
```

```python
def generate_metrics_df():
    # Create a DataFrame
    metrics_df = pd.DataFrame({
        'Accuracy': accuracy_scores.values(),
        'Precision': precision_scores.values(),
        'Recall': recall_scores.values(),
        'F1 Score': f1_scores.values()
    }, index=accuracy_scores.keys())

    return metrics_df.sort_values(by=list(metrics_df.columns),ascending=False)
```
